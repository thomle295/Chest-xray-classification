```python
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
import numpy as np
import matplotlib.pyplot as plt
import os
import tensorflow as tf
import tensorflow.keras as keras
import plotly.graph_objects as go
```

```python
model = keras.models.load_model("./model.h5")
```

```python
PNEUMONIA_VAL_IMAGE = "Dataset/chest_xray/val/PNEUMONIA/person1946_bacteria_4874.jpeg"

img = load_img(PNEUMONIA_VAL_IMAGE, target_size=(256, 256))

plt.imshow(img)
plt.show()

img = img_to_array(img)

predictions = model.predict(np.expand_dims(img, 0))

print("PNEUMONIA(%): " + str(int(predictions[0][1]*100)))
```
